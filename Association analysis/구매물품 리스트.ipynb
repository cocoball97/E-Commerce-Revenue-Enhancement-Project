```python
import pandas as pd
import matplotlib
%matplotlib inline 

import warnings
warnings.filterwarnings('ignore')
matplotlib.rc("font",family = "NaNumGothic")
matplotlib.rc("axes",unicode_minus = False) # 음수표시 
```

```python
df_1= pd.read_csv("../data/actual_purchase.csv",index_col=0)
df_2= pd.read_csv("../data/Member_전처리.csv")
df_3= pd.read_csv("../data/Product_전처리.csv",index_col=0)
df_4= pd.read_csv("../data/날짜별 구매금액.csv")

df_merge = pd.merge(df_1,df_2,on=['회원번호','회원상태'])
df_merge = pd.merge(df_merge,df_3,on='제품번호')

df_merge_20 = df_merge[df_merge['나이']<30]
df_merge_30 = df_merge[(df_merge['나이']>30) & (df_merge['나이']<40)]
df_merge_40 = df_merge[(df_merge['나이']>40) & (df_merge['나이']<50)]
df_merge_50 = df_merge[(df_merge['나이']>50) & (df_merge['나이']<60)]
df_merge_60 = df_merge[(df_merge['나이']>60)]
```

# 연령 별로

```python
# 100개 이상 제품
count_df = df_merge_20['물품명'].value_counts().reset_index()
count_df.columns = ['물품명', '물품명_카운트']
df_merge_20 = pd.merge(df_merge_20, count_df, on='물품명', how='left').query('물품명_카운트 >= 169')

# 같은날 묶어주는 코드
df_merge_20['주문일시'] = pd.to_datetime(df_merge_20['주문일시']) # Convert '주문일시' column to datetime

df_grouped = df_merge_20.groupby(['회원번호', pd.Grouper(key='주문일시', freq='D')]) # Group by '회원번호' and '주문일시' (daily frequency)

df_grouped = df_grouped.agg({'물품명': lambda x: ' $ '.join(x)}) # Concatenate '물품명' values into a single string separated by ','

df_grouped = df_grouped.reset_index() # Reset index to make '회원번호' and '주문일시' into separate columns

df_grouped.rename(columns={'물품명': '전체물품명'}, inplace=True) # Rename column '물품명' to '전체상품명'

# // 두개 하나로
df_grouped['전체물품명'] = df_grouped['전체물품명'].apply(lambda x: x.replace('/ /', '/'))
df_grouped['전체물품명'] = df_grouped['전체물품명'].apply(lambda x: x.replace('//', '/'))

df_grouped['전체물품명'] = df_grouped['전체물품명'].str.strip()
df_grouped['전체물품명'] = df_grouped['전체물품명'].str.replace(' ', '')

#확인
df_grouped[df_grouped['전체물품명'].str.contains(' ')]

# 회원이 몇명인가 확인용도
df_grouped.drop_duplicates(subset=['회원번호'])

df_grouped.to_csv('../data/같이사는물품 20대.csv', encoding="utf-8-sig")
```
