```python
import pandas as pd
import matplotlib
%matplotlib inline 
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
matplotlib.rc("font",family = "Malgun Gothic")
matplotlib.rc("axes",unicode_minus = False) # 음수표시 
```

```python
df= pd.read_csv("../data/02 Member_Data.csv",index_col=0)
```

```python
# 나이 이상치 제거
df = df[df['나이']<90]
```

```python
# 결측치 처리
df['회원상태'].fillna('정상회원',inplace=True)
df['등록카드'].fillna('미기입',inplace=True)
df['결혼'].fillna('미기입',inplace=True)

# 결측치 제거
df.dropna(subset=['성별'],inplace=True)
```

```python
# 탈퇴 정리
df.loc[df['회원상태']=='탈퇴처리중','회원상태'] = '탈퇴'
df.loc[df['회원상태']=='탈퇴신청','회원상태'] = '탈퇴'
```

```python
# 주소 처리
df.loc[df['주소지']=='서울','주소지']='서울특별시'
df.loc[df['주소지']=='강원','주소지']='강원도'
df.loc[df['주소지']=='경기','주소지']='경기도'
df.loc[df['주소지']=='광주','주소지']='광주광역시'
df.loc[df['주소지'].isnull(),'주소지']='미기입'
```
